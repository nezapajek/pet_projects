### Day 7 - No Space Left On Device

```python
# import data
#filename = '07_test.txt'
filename = '07_data.txt'
raw = open(f'./data/{filename}').read()
```

```python
# define data process function
def process(data):
    data = data.split('\n')
    return data

# process data
data = process(raw)
```

```python
from collections import defaultdict

path = []
size_dict = defaultdict(int) 

for line in data:
    words = line.strip().split() #remove leading, trailing spaces & split string
    if words[1] == 'cd':
        if words[2] == '..': #move up a dir
            path.pop() #copy a removed value
        else:
            path.append(words[2]) #add to path
    elif words[1] == 'ls':
        continue
    else:
        try:
            size = int(words[0]) #make size int
            print(path, size)
            for i in range(len(path)+1): #join whole path and dir size
                size_dict['/'.join(path[:i])] += size 
        except:
            pass
```

```python
ans = 0
for key, value in size_dict.items():
    if value <= 100000:
        ans += value
print(ans)
```

### Part 2

```python
all_space = size_dict['/']
```

```python
max_space = 70000000
unused_space = 30000000
available_space = max_space-unused_space
to_delete = all_space-available_space
```

```python
big_vals = []
for key, val in size_dict.items():
    if val >= all_space-available_space:
        big_vals.append(val)
```

```python
res_val = min(big_vals, key=lambda x: abs(to_delete - x))
res_val
```

```python
# alternative
import numpy as np
sorted_size = np.array(sorted(list(size_dict.values())))  # sort sizes and convert to np.array
sorted_size[sorted_size >= to_delete][0]  # get ones that are large enough and take the first one (smallest, cause ordered)
```
