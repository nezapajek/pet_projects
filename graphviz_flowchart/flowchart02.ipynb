```python
import graphviz
```

```python
# start of the object
dot = graphviz.Digraph(comment='flowchart')
dot.attr(newrank='true', compound='true') 

#define styles
input_style = {'shape':'folder','fontsize':'12pt','color':'black','fillcolor':'#377eb880','style':'filled'}
method_style = {'shape':'box','fillcolor':'#ff7f0080', 'style':'filled'}
classifier_style = {'shape':'box', 'fillcolor':'#4daf4a80', 'style':'rounded, filled'}
marker_cluster_style = {'shape':'box', 'color':'#f781bf80', 'fillcolor':'mediumseagreen'}
white_style = {'shape':'plaintext'}
metadata_style = {'shape':'note'}
```

```python
#methods
dot.node('mt1', 'SAMPLING', method_style)
dot.node('mt2', 'DNA ISOLATION', method_style)
dot.node('mt3', 'DNA PREAMPLIFICATION \n EquiPhi 29 polymerase', method_style)
dot.node('mt4', 'PCR', method_style)
dot.node('mt5', 'DNA SEQUENCING \n Illumina MiSeq', method_style)
dot.node('mt6', 'ANALYSIS', method_style)

dot
```

```python
# add edges
dot.edge('mt1','mt2')
dot.edge('mt2','mt3')
dot.edge('mt3','mt4')
dot.edge('mt4', 'mt5')
dot.edge('mt5', 'mt6')
dot
```
